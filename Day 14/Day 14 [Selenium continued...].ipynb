```python
driver = webdriver.Chrome(executable_path='C:\\Users\\Abhijit Gadge\\Downloads\\Selenium Drivers\\chromedriver_win32\\chromedriver.exe')
driver.get("https://www.youtube.com/")
```

```python
trending_link = driver.find_element_by_link_text('Trending')
trending_link.click()
```

```python
html_data = driver.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html_data, 'html.parser')
Title_list = soup.find_all('a', {'id': 'video-title'})
views_list = soup.find_all('span', {'class':'style-scope ytd-video-meta-block'})[::2]
```

```python
log_file = "Trending Videos.txt"
with open(log_file, 'w+', encoding="utf-8") as file:
    for video, views in zip(Title_list, views_list):
        line = " : ".join([video.text, views.text, "\n"])
        file.write(line)
```

```python
with open(log_file, 'r', encoding="utf-8") as file:
    logs = file.read()
    print(logs)
```
