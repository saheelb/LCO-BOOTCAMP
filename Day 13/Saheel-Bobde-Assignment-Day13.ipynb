```python
import requests as rq
from bs4 import BeautifulSoup
price = input("Enter the price : ")
url = "https://www.amazon.in/s?k=smartphones+under+" 
price_url = url + price
head={'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.134 Safari/537.36"}
web_req = rq.get(price_url, headers=head)
web_text = web_req.text

soup_obj = BeautifulSoup(web_text,'html.parser' )
smartphones = soup_obj.findAll('span', { 'class' : 'a-size-base-plus a-color-base a-text-normal'})
cost = soup_obj.findAll('span', {'class':'a-price-whole'} )
for i in zip(smartphones,cost):
    print("Phone: {} and price: {}".format(i[0].get_text(), i[1].get_text()))
```
