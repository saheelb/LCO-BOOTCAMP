```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5,6])
y = np.array([4,7,5,5,8,9])
print( x ,y )
Xm = np.mean(x)
Ym = np.mean(y)
# print(Xm , Ym)

mrows = len(x)
num = 0
denom = 0
c= 0
#y = mx + c
for row in range(mrows):
    num = num + (x[row] - Xm)*(y[row] - Ym)
    denom = denom +  (x[row] - Xm)**2
slope = num/denom
c = Ym - slope*Xm
print(slope , c)
```

```python
plt.plot(x ,y )
plt.plot(Xm , Ym , 'o')
```
