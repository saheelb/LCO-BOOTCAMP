```python
import math
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("123.jpg" , 0)
```

```python
img.shape
```

```python
rows , cols = img.shape
```

```python
img_output = np.zeros(img.shape)
```

```python
for i in range(rows):
    for j in range(cols):
        
        offset_x = int(50.0*math.cos(2*3.14*i / 180))
        offset_y = 0
        
        if j+offset_x < cols:
            img_output[i,j] = img[i , (j+offset_x)]
        else:
            img_output[i,j] = 255
            
        #for drwaing a line acrorss cols at middle 
    
        if (j+offset_x > cols//2) and (j+offset_x < cols//2 + 10 ):
             img_output[i,j] = 255
            
```

```python
img[2,25]
```

```python
cv2.imwrite("123_3.jpg" , img_output)
```

```python
plt.imshow(img_output)
```

```python
for i in range(rows):
    for j in range(cols):
        
        
        offset_x = 0
        offset_y = int(50.0*math.sin(2*3.14*j / 180))
        
        if i+offset_y < rows:
            img_output[i,j] = img[(i+offset_y) ,j ]
        else:
            img_output[i,j] = 255
         
        #for drwaing a line acrorss rows at middle 
        if (i+offset_y > rows//2 ) and (i+offset_y < rows//2 + 10 ):
               img_output[i,j] = 255
            
            
```

```python
cv2.imwrite("123_4.jpg", img_output)
```

```python
plt.imshow(img_output)
```

```python
for i in range(rows):
    for j in range(int(cols/2)):
        img_output[i][j] = 255
        
```
