```python
num_rows , num_cols =img.shape[:2]
```

```python
root_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), 350 , 1.0)
```

```python
root_matrix
```

```python
img_rotation = cv2.warpAffine(img , root_matrix , (num_cols , num_rows ))
```

```python
print(img_rotation)
```

```python
mat.imshow(img_rotation)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as mat
```

```python
img = cv2.imread("22222.jpg", 1)
```

```python
img
```

```python
mat.imshow(img)
```

```python
dest_points = np.float32([[num_cols*0.3,0],[num_cols*0.7, 0],[0,num_rows-1],[num_cols - 1 , num_rows-1]])
```

```python
dest_points
```

```python
src_points = np.float32([[0,0],[num_cols-1,0],[0,num_rows-1],[num_cols-1,num_rows-1]])
```

```python
src_points
```

```python
stretched_matrix = cv2.getPerspectiveTransform(src_points,dest_points )
```

```python
stretched_matrix
```

```python
img_out = cv2.warpPerspective(img ,stretched_matrix, (num_cols,num_rows) )
```

```python
cv2.imwrite("22222.jpg", img_out)
```
