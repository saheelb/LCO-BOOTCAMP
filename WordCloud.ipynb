```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)
import random

def grey_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl({}, {}%, {}%)".format(str(random.randint(0, 150)),str(random.randint(80, 100)), str(random.randint(40, 50)))


def show_wordcloud(data, title = None):
    
#     alice_mask = np.array(Image.open("C:\\Users\\sampr\\Desktop\\about.png"))
    
    wordcloud = WordCloud(
        font_path='C:\\Users\\Saheel Bobde\\OneDrive\\Desktop\\aaa.ttf',
        background_color='black',
        stopwords=stopwords,
        max_words=100,
        max_font_size=40, 
        scale=4,
        height=100,
        width=100,
        mask=None,
        repeat=True,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data)).recolor(color_func=grey_color_func, random_state=3)

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud,
           interpolation="bilinear")
    plt.show()
#     plt.imsave("Hello.png",wordcloud)
    
data = "Saheel Anurag Abhijit LCO Bootcamp will miss you Had fun Here Learned lot of new things "
# show_wordcloud(data)
```
