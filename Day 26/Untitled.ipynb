```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4]).reshape((-1, 1))
y = np.array([9.6, 9.5, 9.04, 8.98]).reshape((-1, 1))
```

```python
marks = LinearRegression().fit(x,y)
intercept = marks.intercept_
coef = marks.coef_
```

```python
y_pred = intercept + coef*x
y_pred
```

```python
sem_5 = marks.predict(np.array([5]).reshape((-1, 1)))
sem_5
```

```python
plt.plot(x, y, "--o")
plt.plot(x, y_pred, "r")
```
