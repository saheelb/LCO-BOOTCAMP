```python
cap = cv2.VideoCapture(0)
ret , frame = cap.read()
gray = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
print(ret)
# print(frame)
# cv2.imshow('frame ', frame)
plt.imshow(gray)
```

```python

cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    c =  cv2.waitKey(1)
    if c==27:
        break
cv2.imshow("Cam 1", cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY))
cap.release()
```
